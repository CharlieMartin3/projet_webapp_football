```python
players = pd.read_csv('players_data.csv')
gk = pd.read_csv('gk_data.csv')
transfermkt = pd.read_csv('transfermkt_url.csv')
```

```python
complete_players = pd.merge(players, transfermkt, left_on='Player', right_on='nom_joueur', how='inner')
complete_gk = pd.merge(gk, transfermkt, left_on='Player', right_on='nom_joueur', how='inner')
```

```python
# Function to convert values to float (in million euros)
def convert_value(value):
    if pd.isna(value):  # Handle NaN values
        return None
    value = value.replace("‚Ç¨", "").replace(",", ".").strip()
    if "mio." in value:
        return float(value.replace("mio.", "").strip())  # Convert "mio." values directly
    elif "K" in value:
        return float(value.replace("K", "").strip()) / 1000  # Convert "K" values to million
    return None
```

```python
complete_players["valeur_marchande"] = complete_players["valeur_marchande"].apply(convert_value)
complete_gk["valeur_marchande"] = complete_gk["valeur_marchande"].apply(convert_value)
```
